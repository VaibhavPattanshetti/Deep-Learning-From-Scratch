```python
import numpy as np
import pandas as pd
```

```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_index
```

```python
tokenizer.document_count  #print number of samples
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences

sequences = pad_sequences(sequences, padding='post')
sequences
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
print(X_train.shape)
print(X_test.shape)
```

```python
print(len(X_train[0]))
print(len(X_train[1]))
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape #all sample are now same shape
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False))  # 50 -> timesteps 1 -> no. of inputs
model.add(Dense(1,activation='sigmoid'))

model.summary()

# 1. return_sequences=False -> Output only the last hidden state
# 2. return_sequences=True -> Output hidden state at every time step
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test))
```
